```python
class ConfigureChannels():
    def __init__(self):
        super().__init__()
        self._pulser_channels = {
            'ch_aom': 0, # output channel 0
            'ch_switch': 1, # output channel 1
            'ch_daq': 2, # output channel 2
            'ch_sync': 3 # output channel 3
        }
        self._timetagger_channels = {
            'click_channel': 1,
            'start_channel':2,
            'next_channel':-2,
            'sync_channel':tt.CHANNEL_UNUSED,
        }
    @property
    def pulser_channels(self):
        return self._pulser_channels
    @property
    def timetagger_channels(self):
        return self._timetagger_channels  
_configure_channels = ConfigureChannels()
pulser = _configure_channels.pulser_channels
pulser
```

```python
import numpy as np

# 原始列表
original_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# 转换为NumPy数组
numpy_array = np.array(original_list)

# 分割成子数组
subarrays = np.split(numpy_array, [3, 6])

# 按照要求进行组合
result = np.concatenate(subarrays)

# 将结果转换为列表
result_list = result.tolist()
print(subarrays)
print(result_list)
```
